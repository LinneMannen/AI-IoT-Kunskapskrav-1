```python
# Importer
import numpy as np
import polars as pl
from dotenv import load_dotenv
from google.genai import types
from google import genai
import os
from embedding_service import EmbeddingService
from vector_store import VectorStore
```

```python
store = VectorStore("embeddings_v2.parquet")
```

```python
## Snabb check
print(store.df.head())
print(store.embedding_matrix.shape)
```

```python
load_dotenv()

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

service = EmbeddingService(client, types)

query = "Hur sätter man segel?"

query_embedding = service.embed(query)

results = store.search(query_embedding, k=10)

results
```
