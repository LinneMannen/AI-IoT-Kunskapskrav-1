1. Antag att du ska implementera en AI-baserad chattbot för kundsupport. Beskriv
arbetsprocessen utifrån design thinking-modellens olika faser och förklara syftet
med varje fas.

##### Empathize 
Här går det ut på att förstå användaren och samla insikter kring deras behov.
Fokus ligger på att förstå verksamheten ofta igenom intervjuer och/eller andra observationer.
Detta bidrar till att behovet sätts utifrån verksamhets faktiska behov och inte utifrån vad utvecklaren eller verksamheten tror att de behöver. Detta minskar risken att bygga en lösning för fel problem och skapar bra underlag för nästa steg.

##### Define
I denna del använder vi insikterna från Empathize delen för att definiera problemen på ett tydligt sätt.
På så sätt skapar vi ett eller flera tydliga problem att arbeta mot i projektet.
Det skulle kunna vara att hjälpa kunder att hitta information om exempelvis returpolicy eller orderstatus utan att behöva ringa någon. Syftet är att se till att vi arbetar mot rätt problem.

##### Ideate
Här brainstormar vi fram olika lösningar på problem som vi definierat. Fokus ligger på att hitta så många och kreativa lösningar som möjligt, snarare än att försöka hitta de bästa lösningarna som löser alla våra problem.
Exempel på lösningar skulle kunna vara att göra en RAG chattbot eller kanske en hybridlösning där användaren först kan beskriva sina problem och fylla i en "mall" som senare gör det lättare för kundsupport att hjälpa kunden på ett effektivt sätt.

##### Prototype 
I denna del byggs prototyper av lösningarna för att testa våra ideer. Prototyperna ska vara billiga och snabba att skapa.
Detta steg är mer som ett proof of concept än at faktiskt göra en färdig lösning. Här är fokus på att testa våra ideer inte skapa en produkt.

##### Test
Här testar vi prototyperna på ett så verkligt sätt som möjligt och samlar feedback för nya insikter. Denna del är alltså inte bara till för att utvärdera lösningen utan också för att få ökad förståelse av användarens behov. Prototype och Test delarna arbetar man ofta iterativt med för att på så sätt utveckla ideerna och lösningarna till en så bra produkt och lösning som möjligt.


2. Beskriv skillnaden mellan ett teknikdrivet och ett användarcentrerat
utvecklingssätt.

##### Teknikdrivet utvecklingssätt
Utgår från vad som är möjligt att bygga. Fokus ligger på tekniken exempelvis AI, IoT eller andra program och att man bygger lösningen på detta. Syftet är att använda ny teknik och effektivisera processer, risken är att man utvecklar något som inte löser ett verkligt problem. Ett exempel skulle kunna vara att bygga en chattbot just för att det är populärt men eftersom att fokus inte lagts på användaren så löses inget verkligt behov.

##### Användarcentrerat utvecklingssätt
Utgår istället från användarens behov och problem. Använder design-thinking processerna för att förstå användaren och utveckla en lösning som verkligen behövs. Syftet är att skapa verkligt värde för användaren.


3. Ge exempel på hur IoT-data kan användas inom Business Intelligence för att
stödja affärsbeslut. Beskriv vilken typ av data som samlas in och hur den kan
analyseras.

Ett företag skulle kunna använda sig av IoT-sensorer i exempelvis lager, fabriker eller stora butiker för att få koll på lager och flöden i realtid. Detta skulle kunna optimera lagerhållning, planera inköp och service för att på så sätt förhindra dyra driftstopp och effektivisera logistik. 

Data som samlas in skulle kunna vara exempelvis lagerstatus, flöde av varor eller produktionslinjr eller annan maskindata.

Datan som samlats in skulle kunna visualiseras i realtidsdashbords där vi också kan prediktera framtida utfall och försöka upptäcka problem eller möjligheter. Allt detta i realtid för att få en bättre aktuell bild av verksamheten.


4. Hur skiljer sig IoT-data från traditionell affärsdata och vilka konsekvenser får detta
för BI-arbete?

##### IoT-data
- Samlas in kontinuerligt i realtid
- Ofta stora datamängder
- Kan vara ostrukturerad eller semistrukturerad exempelvis JSON eller loggar.
- Handlar ofta om fysiska händelser exempelvis temperatur eller rörelse 

##### Traditionell affärsdata
- Samlas in i batch ofta månadsvis veckovis eller dagligen. 
- Är mer strukturerad i till exempel tabeller i databaser.
- Handlar om affärshändelser som exempelvis ordrar, fakturor eller kunder.

Konsekvenserna som IoT-data medför är flera. IoT-data kräver BI-lösningar som kan hantera streamingdata i realtid vilket skapar större krav på plattform och modellering. Eftersom IoT data också ofta är i stora mängder så ökar kravet även det kravet på städning och hantering av datan. Det kanske inte hinns med att gå igenom ett ordentligt ETL flöde utan får istället städas snabbt med schemalagda skript eller liknande. Detta i sin tur möjliggör prediktiva analyser och realtidsbeslut som inte är lika lätt i traditionell BI.  

5. Hur kan AI möjliggöra nya affärsmodeller snarare än enbart effektivisering?
AI har i stor utsträckning använts för effektivisering men det finns även många nya affärsmodeller som möjligjorts. 

Som vi pratade om under en föreläsning så kan det till exempel handla om riktade erbjudanden via chattbotar som anpassar sig efter användarens behov. Detta görs redan med data från nätet men en AI chattbot möjliggör en väldigt mycket mer personaliserad profil.

 Ett annat exempel som passar mig med en bakgrund som hantverkare skulle kunna vara en underhållsplan för fastigheter med hjälp av AI. Där man enkelt matar in fastighetens data så som energiförbrukning, byggår, kvm, takstorlek osv. För att på så sätt ha en AI förvaltare av fastigheterna som inte bara effektiviserar utan även sparar pengar, fungerar som bollplank med hjälp av kraftfulla LLMer, planerar underhåll och identifierar risker i förväg.

 Detta möjliggör en affärsmodell som är enkel att skala och billig för fastighetsägare att implementera. Allt detta som ett abonnemang istället för som en tjänst där en anställd annars hade behövts.
